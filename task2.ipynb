## Intensity clipping

First we define some fixed thresholds $T_1$, $T_2$:

```python
T1 = 0.31
T2 = 0.59
```

### Using `ndarray.clip` method

**<a id="#">Task 2.3.1.</a>** Enter your solution below:

```python
copy = img.copy()
copy.flatten()
ccopy = copy.clip(T1,T2)
plt.hist(ccopy.flatten())
```

### Not using the `ndarray.clip` method

**<a id="#">Task 2.3.2 (a).</a>** Enter your solution below:

```python
h, w = img.shape

for i in range(h):
    for j in range(w):
        if img[i,j] < T1:
            copy[i,j] = T1
        elif img[i,j] > T2:
            copy[i,j] = T2
        else:
            copy[i,j] = copy[i,j]
plt.hist(copy.flatten())
```
